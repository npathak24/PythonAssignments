```python
print("Task-1 Create a Jupyter Notebook to present your analysis and perform an exploratory data analysis (EDA) on flights.csv file")

#import flights.csv file
import pandas as pd

df = pd.read_csv(r'C:\Users\Naman\Documents\New folder\flights.csv')

# EDA
# 1.Print first five rows of data
print("\nFirst five rows: ")
print(df.head())

# 2. Check for missing values in data
print("\nMissing values: ")
print(df.isnull().sum())

# 3. Check data type of each column
print("\nData type of each column: ")
print(df.dtypes)

# 4. Print no. of rows and columns
print("\nNo. of rows: ", df.shape[0])
print("No. of columns: ", df.shape[1])

# 5. Summary statistics of the numerical columns
print("\nstatistics of numerical columns: ")
print(df.describe())
```

```python
print("Task-2 Create at least 5 different views (reports) to the team using the imported dataset.")
print("The reports should be composed of a description, charts and tables based on some selected columns.")
print("These reports can give some insights to the team about the airport operation and potential optimizations")

import matplotlib.pyplot as plt
import seaborn as sns

# Report - 1
print("\n1. Departure Delay Distribution")
print("\nThis report provides an overview of the distribution of departure delays." "It includes a histogram showing the"
      "distribution of departure delays, as well as a table listing the statistics of departure delays. This report can"
      "help the team understand how often flight departures experience delays and optimize the operations accordingly.")

# Create a histogram showing the distribution of departure delays
print("\n Show a histogram showing the distribution of departure delays and a table showing statistics of departure delays: ")
plt.hist(df['dep_delay'], bins=20)
plt.xlabel('Departure Delay (min)')
plt.ylabel('Frequency')
plt.show()

# Create a table showing the statistics of departure delays
print(df['dep_delay'].describe())

# Report - 2
print("\n2. Average Arrival Delay by Carrier")
print("\nThis report provides an overview of the average arrival delay for each carrier. It includes a bar chart" 
      "showing the average arrival delay for each carrier, as well as a table listing the carriers and their average" 
      "arrival delay. This report can help the team identify carriers that may be experiencing delays more frequently" 
      "and help the team to optimize the operations.")
print("\nCalculate average and show a bar chart and a table showing the average arrival delay by carrier: ")
# Group the data by carrier and calculate the average arrival delay
carrier_avg_arrival_delay = df.groupby('carrier')['arr_delay'].mean()

# Create a bar chart showing the average arrival delay by carrier
carrier_avg_arrival_delay.plot(kind='bar')
plt.xlabel('Carrier')
plt.ylabel('Average Arrival Delay (min)')
plt.show()

# Create a table showing the average arrival delay by carrier
print(carrier_avg_arrival_delay)

# Report - 3
print("\n3. Number of Flights by Carrier")
print("\nThis report provides an overview the number of flights operated by a particular airline. It includes a pie chart"
      "showing number of flights by airline and also a table showing the same.")
print("\n Display pie chart and table showing number of flights by carrier: \n")
# Group the data by destination and count the number of flights
flights_by_carrier = df.groupby('carrier').size()

# Create a pie chart to visualize the number of flights by destination
flights_by_carrier.plot(kind='pie')
plt.show()

# Create a table showing the number of flights by month
print(flights_by_carrier)

# Report - 4
print("\n4. Number of Flights by Month")
print("\nThis report provides an overview of number flights operated within a month. It includes a line chart showing number"
     "of flights by month and also a table showing the same.")
print("\n Display line chart and table showing number of flights by month: \n")
# Group the data by month and count the number of flights
flights_by_month = df.groupby(df['month']).size()

# Create a line chart to visualize the number of flights by month
flights_by_month.plot(kind='line')
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()

# Create a table showing the number of flights by month
print(flights_by_month)

# Report - 5
print("\n5. Early Arrival Flights by Carrier ")
print("\n This report provides an overview of flights arrived early based on airlines. It includes a"
     "bar graph and table showing percentage of number of flights arrived early")

print("\nDisplay percentage of early arrived flights in bar graph and table: \n")
#Create a column Early to include flights arrived early
df['Early'] = df['arr_delay'] < 0

#Calculate percenatge of early arrived flights and display in bar graph
early_flights = df.groupby(['carrier', 'Early']).size().unstack()
early_pct = early_flights[True] / (early_flights[True] + early_flights[False]) * 100
early_pct.plot(kind='bar')
plt.xlabel('Airline Carrier')
plt.ylabel('Percentage of Early Flights')
plt.title('Percentage of Flight Arrived Early by Airline Carrier')
plt.show()

#Create a table showing percentage of early arrived flights
print(early_pct)

```

```python
print("Task-3 Create a new dataset named (pop_data) with flight data from airlines UA (United Airlines) and DL (Delta Airlines)."
      " The data set must contain only two columns, company name and delayed arrival flights. The data must be extracted from"
      "dataset flights.csv to build the pop_data dataset. The dataset should be limited to no more than 20,000 rows per airlines")
# Create a new dataframe with only data from UA and DL airlines
pop_data = df[(df['carrier'] == 'UA') | (df['carrier'] == 'DL')]

# Limit the dataframe to no more than 20,000 rows per airlines
pop_data = pop_data.groupby('carrier').head(20000)

# Create a new column for the company name
pop_data['company'] = pop_data['carrier'].replace({'UA':'United Airlines', 'DL':'Delta Airlines'})

# Keep only the company name and delayed arrival flights columns
pop_data = pop_data[['company', 'arr_delay']]

# Rename the arr_delay column to delayed_arrival_flights
pop_data = pop_data.rename(columns={'arr_delay':'delayed_arrival_flights'})

print("\nDisplay new dataset with columns 'Company Name' and 'Delayed Arrival Flights' having only UA and DL information:  \n")
print(pop_data)
```

```python
print("Task-4 Create two new datasets (dl and ua) of 1000 observations each from the pop_data dataset only with data from"
      "the DL (Delta Airlines) for dl and only data from UA (United Airlines) for ua . Tip: Include a column called"
      "sample_id populated with number 1 for the first sample and 2 for the second sample")
print("\nDisplay DL dataset with 1000 observations: \n")
# Create the "dl" dataset with 1000 observations of Delta Airlines data
dl = pop_data[pop_data['company'] == 'Delta Airlines'].sample(1000)
dl['sample_id'] = 1
print(dl)

print("\nDisplay UA dataset with 1000 observations: \n")
# Create the "ua" dataset with 1000 observations of United Airlines data
ua = pop_data[pop_data['company'] == 'United Airlines'].sample(1000)
ua['sample_id'] = 2
print(ua)
```

```python
print("Task-5 Create a new dataset (samples) containing the data of the 2 samples created in the previous item to be used in future analysis")
print("\nAdd and display new sample set having information of above DL and UA sample sets: \n")
# Concatenate the "dl" and "ua" datasets
samples = pd.concat([dl, ua], axis=0)
print(samples)
```

```python
print("Task-6 For each sample (dl and ua) calculate de standard error and the mean.")
print("\nDisplay mean and standard error for each sample('dl' and 'ua'): \n")
# calculate the standard error for the 'arr_delay' column in the 'dl' dataset
dl_std_err = dl['delayed_arrival_flights'].sem()

# calculate the mean for the 'arr_delay' column in the 'dl' dataset
dl_mean = dl['delayed_arrival_flights'].mean()

# calculate the standard error for the 'arr_delay' column in the 'ua' dataset
ua_std_err = ua['delayed_arrival_flights'].sem()

# calculate the mean for the 'arr_delay' column in the 'ua' dataset
ua_mean = ua['delayed_arrival_flights'].mean()

print("\nDelta Airlines (DL) standard error: ", dl_std_err)
print("Delta Airlines (DL) mean: ", dl_mean)

print("\nUnited Airlines (UA) standard error: ", ua_std_err)
print("United Airlines (UA) mean: ", ua_mean)
```

```python
print("Task-7 For each mean calculated before we need to define the confidence intervals in this case 95% confidence interval."
      " It means calculate lower and upper values")
print("\nDisplay lower and upper values of DL and UA dataset w.r.t 95% confidence interval: ")
from scipy.stats import t
import numpy as np
import warnings
warnings.simplefilter('ignore')

# calculate the standard deviation of the 'arr_delay' column in the 'dl' dataset
dl_std = dl['delayed_arrival_flights'].std()

# calculate the sample size of the 'dl' dataset
dl_sample_size = dl.shape[0]

# calculate the 95% confidence interval for the mean of the 'arr_delay' column in the 'dl' dataset
dl_conf_int = t.interval(alpha = 0.95,              # Confidence level
              df= dl_sample_size-1,                # Degrees of freedom
              loc = dl_mean,                       # Sample mean
              scale = dl_std/np.sqrt(dl_sample_size))  # Standard deviation estimate

print("\nDelta Airlines (DL) lower and upper values are: ", dl_conf_int)

# calculate the standard deviation of the 'arr_delay' column in the 'ua' dataset
ua_std = ua['delayed_arrival_flights'].std()

# calculate the sample size of the 'ua' dataset
ua_sample_size = ua.shape[0]

# calculate the 95% confidence interval for the mean of the 'arr_delay' column in the 'ua' dataset
ua_conf_int = t.interval(alpha = 0.95,              # Confidence level
              df= ua_sample_size-1,                # Degrees of freedom
              loc = ua_mean,                       # Sample mean
              scale = ua_std/np.sqrt(ua_sample_size))  # Standard deviation estimate

print("\nUnited Airlines (UA) lower and upper values are: ", ua_conf_int)
```

```python
print("Task-8 After these set of previous calculations it was requested that you take the T-TEST concept and apply it on your"
      " 2 groups of data (means) to let your team knows if Delta Airlines (DL) flights are delayed more than United Airlines (UA) flights ?")
print("\nT-Test to determine whether Delta Airline more delayed than United Airline: \n")
import warnings
from scipy import stats

warnings.simplefilter('ignore')

alpha=0.05 #critical statistical significance

t, p_value = stats.ttest_ind_from_stats(dl_mean, dl_std, dl_sample_size, ua_mean, ua_std, ua_sample_size, equal_var=False)

# print the t value
print("t value: ", t)
# print the p-value
print("p-value: ", p_value)

if p_value <= alpha :
    print("\nT-Test concludes that we reject the null hypothesis and there is significant difference between mean of two groups."
          "  So, Delta Airline are more delayed that United Airline ")
else :
    print("\nT-Test concludes that we can't reject null hypothesis and there is not much significant difference between mean"
          "of two groups. So, we can't draw any significant conclusion ")
```
