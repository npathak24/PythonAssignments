```python
# Python program to create console based application for Online Banking
# Create Bank_Account class
import random
import re
class Bank_Account:
    my_dict = {}

# Initialisation
    def __init__(self):
        self.balance=0
        self.account_number={}
        print("Hello!!! Welcome to the Online Banking Application")


# Function for bank account creation and one customer can create many accounts
# Function to create a bank account        
    def create_account(self):
        pat = "^[a-zA-Z0-9-_]+@[a-zA-Z0-9]+\.[a-z]{1,3}$"
        name=str(input("Enter your name: "))
        self.name=name
        email=str(input("Enter your email id: "))
        if re.match(pat,email):
            self.email=email
        else:
            email=str(input("\nPlease enter your email in 'name@domain.ext' format: "))
        self.email=email
        my_list = [self.name]
        my_list.append(self.email)

# Generate account number and store them in dictionary so that one account is linked to one customer only
        account_number=random.randint(1000000000, 9999999999)
        print("Welcome!",name,",your account has been created successfully and your account number is: ", account_number)
        self.my_dict[account_number]=my_list

        
# Function to deposit amount
    def deposit(self):
        account=int(input("Enter your account number: "))
        if account in self.my_dict:
            print("Hello!", self.my_dict[account])
            amount=float(input("Please enter amount to be Deposited: "))
            self.balance += amount
            print("\n Amount Deposited: $",amount)
            print("\n Net Available Balance= $",self.balance)
        else :
            print("\nInvalid Account Number")


#Functions for withdraw amount and transfer amount and negative balance is not acceptable
# Function to withdraw amount
    def withdraw(self):
        account=int(input("Enter your account number: "))
        if account in self.my_dict:
            print(self.my_dict[account])
            amount = float(input("Enter amount to be Withdrawn: "))
            if self.balance>=amount:
                self.balance-=amount
                print("\n You Withdrew: $",amount)
                print("\n Net Available Balance= $",self.balance)
            else:
                print("\n Insufficient balance ")
        else:
            print("\nInvalid Account Number")

# Function to transfer amount from one account to another            
    def transfer_money(self):
        account=int(input("Enter your account number: "))
        if account in self.my_dict:
            print(self.my_dict[account])
            to_account=int(input("Enter the account number to which you want to transfer amount: "))
            if len(str(to_account))!=10:
                to_account=int(input("Account number must be 10 digits(numbers only): "))
                
            if to_account==account:
                to_account=int(input("Both account cannot be same, please enter correct account number: "))
            amount=float(input("Enter amount to be transferred: "))
            if self.balance>=amount:
                self.balance-=amount
                print("$",amount, "Transferred")
                print("\n Net Available Balance= $",self.balance)
            else:
                print("Insufficient Balance")
        else:
            print("\nInvalid Account Number")
    

# Funtion to show bank balance
    def account_balance(self):
        print("\n Your Net Available Balance is: $",self.balance)

            
# creating an object of class
s = Bank_Account()


# Calling functions in a while loop
n=1
while(n > 0 ):
    print("\n\nOptions: ")
    print("1. Create Account")
    print("2. Deposit Amount")
    print("3. Withdraw Amount")
    print("4. Transfer Amount")
    print("5. Account balance")
    print("6. Quit")
    x=int(input("\nEnter your choice: "))
    
# Calling functions with that class object
    if x==1:
        s.create_account()
        choice=input("\nDo you want to continue(y/n)? ")
        if choice=='n':
            print("\nThank you for using online banking application")
            break
        elif choice != 'y':
            print("\nWrong input")
            choice=input("\nPlease enter with y or n: ")
    elif x==2:
        s.deposit()
        choice=input("\nDo you want to continue(y/n)? ")
        if choice=='n':
            print("\nThank you for using online banking application")
            break
        elif choice != 'y':
            print("\nWrong input")
            choice=input("\nPlease enter with y or n: ")
    elif x==3:
        s.withdraw()
        choice=input("\nDo you want to continue(y/n)? ")
        if choice=='n':
            print("\nThank you for using online banking application")
            break
        elif choice != 'y':
            print("\nWrong input")
            choice=input("\nPlease enter with y or n: ")
    elif x==4:
        s.transfer_money()
        choice=input("\nDo you want to continue(y/n)? ")
        if choice=='n':
            print("\nThank you for using online banking application")
            break
        elif choice != 'y':
            print("\nWrong input")
            choice=input("\nPlease enter with y or n: ")
    elif x==5:
        s.account_balance()
        choice=input("\nDo you want to continue(y/n)? ")
        if choice=='n':
            print("\nThank you for using online banking application")
            break
        elif choice != 'y':
            print("\nWrong input")
            choice=input("\nPlease enter with y or n: ")
    elif x==6:
        print("\nThank you for using online banking application")
        break
    else :
        print("Wrong Input")
        print("Please enter number from the available options")
    
    n+=1
    
# End of program
```
